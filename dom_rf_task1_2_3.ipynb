```python
# перевод объектов в строковый формат, для сохранения в базе данных

def handle_df_for_db(df):
  problematic_columns = []
  for column in df.columns:
      if df[column].dtype not in ['int64', 'float64', 'datetime64[ns]']:
          problematic_columns.append(column)

          try:
              df[column] = df[column].astype('str')
          except:
            print(column)
  return df
```

```python
class DbRepository:

  def __init__(self, table_name):
    self._db_path = r'drive/MyDrive/buildings_db.db'
    self._table = table_name
    self._operation_type = 'append'

  def save_df(self, df):
    with sqlite3.connect(self._db_path) as con:
      df.to_sql(self._table, con, index=False, if_exists="replace")

  def get_df(self):
    query = f'select * from {self._table}'
    with sqlite3.connect(self._db_path) as con:
      return pd.read_sql(query, con)

  @property
  def db_path(self):
    return self._db_path

  @property
  def db_table(self):
    return self._table
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display
```

```python
def get_feature_selector(features):

  feature_selector = widgets.Dropdown(
      options=features,
      value=features[0],
      description='Выберите показатель'
  )

  return feature_selector
```

```python
def plot_hist_graph(selected_feature):
    # plt.figure(1, figsize=(15, 15))

    fig, ax = plt.subplots(figsize=(15, 15))

    mean_price_regions.sort_values(selected_feature, inplace=True, ascending=False)
    ax = sns.barplot(y=mean_price_regions.index, x=selected_feature, data=mean_price_regions)
    ax.bar_label(ax.containers [ 0 ])

    plt.title(selected_feature)

    plt.show()
```

```python
# гистограмма по данным
plot_names = mean_price_regions.columns.to_list()
feature_selector = get_feature_selector(plot_names)
interact(plot_hist_graph, selected_feature=feature_selector);
```

```python
# класс извлечения данных с дом рф

class DomRfParser:

  def __init__(self):
      self._offset_ = 1
      self._index = 0
      self._error_request_obj = 0

  @property
  def error_request_obj(self):
    return self._error_request_obj

  def get_object_ids(self, objects_data):
    obj_ids = []

    for object_data in tqdm(objects_data):
      obj_id = object_data.get('objId')
      if obj_id:
        obj_ids.append(obj_id)

    if len(obj_ids) == len(all_objects_data):
      print(f'получение id по объектам завершилось успешно')
    else:
      print(f'id получены не по всем объектам!')

    return obj_ids

  def fetch_building_objects(self, result_limit_=1000, status_object_=0):
    all_objects_data = []

    print(f'скачивание данных объектам строительства со статусом {status_object_}')

    while True:
      url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={self._offset_}&limit={result_limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus={status_object_}'

      res = requests.get(url)

      data = res.json()
      objects_data = data.get('data').get('list')

      all_objects_data.extend(objects_data)

      print(self._index)

      if(len(objects_data) == 0):
        print('скачивание данных по объектам завершено')
        print(f'всего скачано {len(all_objects_data)} объектов')
        self._index = 0
        self._offset = 1

        return all_objects_data

      self._offset += self._limit
      self._index += 1

  def fetch_details_object(self, obj_id):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{obj_id}'

    res = requests.get(url)

    if res.status_code == 200:
        return res.json().get('data')
    else:
        error_request_obj.append(obj_id)
        print(f"Ошибка при запросе для obj_id={obj_id}. HTTP статус: {res.status_code}")
        return None

  def fetch_details_objects_parallel(self, obj_ids, workers_count=30):

    print('старт извлечения данных по объектам')

    with ProcessPoolExecutor(max_workers=workers_count) as executor:
        results = list(tqdm(executor.map(self.fetch_more_data_object, obj_ids), total=len(obj_ids)))

    if len(obj_ids) == len(results):
      print('извлечение детализированных данных по объектам завершено успешно')
    else:
      print(f'данные извлечены не по всем объектам! в error_request_obj - {len(self._error_request_obj)} id объектов')

    return results

  def fetch_result_df(self):
    objects = self.fetch_building_objects()
    object_ids = self.get_object_ids(objects)
    detail_data = self.fetch_details_objects_parallel()

    if len(object_ids) > len(detail_data):
      if len(self.error_request_obj) != 0:
        for error_id in self.error_request_obj:
          detail_data.extend(self.fetch_details_object(error_id))

    return pd.json_normalize(detail_data)
```

```python
# классы сохранения данных

from abc import ABCMeta, abstractmethod, abstractproperty
import os

class DataSourcePdAdapter():
    __metaclass__ = ABCMeta

    def __init__(self, directory_path, file_name):
      self._directory_path = directory_path
      self._file_name = file_name
      self._extension = ''

      self._check_directory_path()
      self._check_filename()

    def _get_full_path(self):
        path = os.path.join(self._directory_path, f'{self._file_name}.{self._extension}')

        return path

    def _check_directory_path(self):
      if self._directory_path == '':
        raise Exception(f'directory_path не может быть пустой строкой!')

      if not os.path.exists(self._directory_path):
          os.mkdir(self._directory_path)

      if not os.path.exists(self._directory_path):
        raise Exception(f'Ошибка при создании каталога {self._directory_path}')

    def _check_filename(self):
      if self._file_name == '':
        raise Exception(f'file_name не может быть пустой строкой!')

    @abstractmethod
    def write(self, df):
      pass

    @abstractmethod
    def _read(self):
      pass

class ExcelPdAdapter(DataSourcePdAdapter):

    def __init__(self, directory_path, file_name, sheet=0, header=0):
      super().__init__(directory_path, file_name)

      self._sheet = sheet
      self._header = header
      self._extension = 'xlsx'

    def write(self, df):
      df.to_excel(self._get_full_path(), index=False)

    def read(self):
      pd.read_excel(self._get_full_path(), sheet_name=self._sheet, header=self._header)


class CsvPdAdapter(DataSourcePdAdapter):

    def __init__(self, directory_path, file_name, sep=','):
      super().__init__(directory_path, file_name)
      self._sep = sep
      self._extension = 'csv'

    def write(self, df):
      df.to_csv(self._get_full_path(), index=False)

    def read(self):
      return pd.read_csv(self._get_full_path(), sep=self._sep)

class PicklePdAdapter(DataSourcePdAdapter):
    def __init__(self, directory_path, file_name):
      super().__init__(directory_path, file_name)
      self._extension = 'pkl'

    def write(self, df):
      df.to_pickle(self._get_full_path())

    def read(self):
      return pd.read_pickle(self._get_full_path())

class SqlLitePdAdapter(DataSourcePdAdapter):
    def __init__(self, directory_path, file_name, table_name):
      super().__init__(directory_path, file_name)
      self._table = table_name
      self._operation_type = 'append'
      self._extension = 'db'

    def write(self, df):
      with sqlite3.connect(self._get_full_path()) as con:
        df.to_sql(self._table, con, index=False, if_exists="replace")

    def read(self):
      query = f'select * from {self._table}'
      with sqlite3.connect(self._get_full_path()) as con:
        return pd.read_sql(query, con)

class Repository:

  def read(self, pd_adapter):
    return pd_adapter.read()

  def write(self, pd_adapters_list, df):
    if len(pd_adapters_list) == 0 or type(pd_adapters_list) != list:
      raise Exception('Некорректный параметр pd_adapters_list, дожен быть не пустой список!')

    for pd_adapter in pd_adapters_list:
      pd_adapter.write(df)

    print('данные сохранены')
```

```python
# класс предобработки данных в датафрейме

class DfProcessor:

  def __init__(self):
    pass

  def rename_columns(self, df, column_rename, new_names):
    df[column_rename] = df[column_rename].map(new_names)
    return df

  def filter_target_columns(self, df, target_columns):
    new_df = df[[col for col in df.columns if col in target_columns]]

    if not (new_df.shape[1] == len(target_columns)):
      raise Exception('Ошибка при фильтре требуемых столбцов!')

    return new_df

  def preprocess_nan(self, df):
    new_df = df.replace('nan', pd.NaT)
    return new_df

  def get_na_list_columns(self, df):
    return df.columns[df.isna().any()].to_list()

  def drop_columns(self, df, droped_columns):
    new_df = df.drop(droped_columns, axis=1)

    if not df.shape[1] == new_df.shape[1] + len(droped_columns):
      raise Exception('Ошибка при удалении колонок!')

    return new_df

  def get_na_columns_by_percent(self, df, percent):

    if percent > 1 and percent < 0:
      raise Exception('Некорректное значение percent, должно быть меньше 1 и больше 0!')

    nan_percentage = df.isna().mean()
    columns_with_high_nan_percentage = nan_percentage[nan_percentage > percent].index

    return list(columns_with_high_nan_percentage)


  def fill_na_columns(self, df, columns, new_value):
    df[columns] = df[columns].fillna(value=new_value)
    return df


class BuildingsDfProcessing:

  def __init__(self, df, df_processor):
    self._initial_df = df
    self._clear_df = None
    self._regions_df = None

    self._df_processor = df_processor

    self._region_column = 'region'
    self._price_column = 'objPriceAvg'

    self._target_columns = [
        'id', self._region_column, 'objElemLivingCnt', 'objReady100PercDt',
       'wallMaterialShortDesc', 'objLkFinishTypeDesc', 'objElemParkingCnt',
       'objSquareLiving', 'objCompensFundFlg', 'objProblemFlg',
       'objLkClassDesc', 'objGuarantyEscrowFlg', 'objectType', 'objFloorCnt',
       'objFlatCnt', 'objNonlivElemCnt', 'isAvailableWantLiveHere',
       'objInfrstrBicycleLaneFlg', 'objInfrstrPlaygrndCnt',
       'objInfrstrSportGroundCnt', 'objInfrstrTrashAreaCnt',
       'objInfrstrObjPrkngCnt', 'objInfrstrNotObjPrkngCnt',
       'objInfrstrRampFlg', 'objInfrstrCurbLoweringFlg',
       'objElevatorWheelchairCnt', 'objElevatorPassengerCnt',
       'objElevatorCargoCnt', 'objElevatorCargoPassengerCnt', self._price_column,
       'objGreenHouseFlg', 'objEnergyEffShortDesc', 'developer.devId',
       'developer.devFullCleanNm', 'developer.problObjCnt',
       'developer.buildObjCnt', 'developer.comissObjCnt',
       'developer.fundGuarantyFlg', 'developer.govFundFlg'
       ]

    self._process_df()

  @property
  def processed_df(self):
    return self._clear_df

  def _process_df(self):
    regions_names = df.groupby([self._region_column])['developer.regRegionDesc'].first()
    self._initial_df = self._df_processor.rename_columns(self._initial_df, self._region_column, regions_names)

    self._initial_df = self._df_processor.filter_target_columns(self._initial_df, self._target_columns)

    self._initial_df = self._df_processor.preprocess_nan(self._initial_df)

    # удаляем колонки где na значений больше 80%
    eihteen_percent_na_columns = self._df_processor.get_na_columns_by_percent(self._initial_df, 0.8)
    if len(eihteen_percent_na_columns) != 0:
      self._initial_df = self._df_processor.drop_columns(self._initial_df, eihteen_percent_na_columns)

    # заполняем нулями пропуски, где меньше 3 процентов
    zero_na_columns = self._initial_df.isna().mean().loc[lambda x: (0 < x) & (x < 0.03)].keys()
    if len(zero_na_columns != 0):
       self._df_processor.fill_na_columns(self._initial_df, zero_na_columns, 0)

    # удаляем записи регионов, по которым во всем регионе не указана цена
    self._regions_df = self._initial_df.groupby([self._region_column])
    none_price_regions = self._get_none_price_regions()
    self._initial_df = self._filter_strings_by_regions(self._initial_df, none_price_regions)

    # заполняем цену средним значением по региону
    self._initial_df[self._price_column] = self._initial_df.groupby([self._region_column])[self._price_column].apply(lambda x: x.fillna(x.mean()))

    # приводим значения к необходимому типу данных
    self._set_types_columns(self._initial_df)

    self._clear_df = self._initial_df.reset_index(drop=True)


  def _get_none_price_regions(self):
    return self._regions_df[self._price_column].mean().isna().loc[lambda x: x == True].keys().to_list()

  def _filter_strings_by_regions(self, df, unnecessary_regions):
    count = df[df[self._region_column].isin(unnecessary_regions)][self._price_column].shape[0]

    new_df = df[~df[self._region_column].isin(unnecessary_regions)]

    if not new_df.shape[0] + count == df.shape[0]:
      raise Exception('Ошибка в фильтре строк регионов!')

    return new_df

  def _as_type(self, df, columns, _type):
    for column in columns:
      df[column] = df[column].astype(_type, errors='ignore')

    return df

  def _set_types_columns(self, df):
    int_type_columns = ['objElemLivingCnt', 'objElemParkingCnt', 'objFloorCnt', 'objFlatCnt',
                    'objNonlivElemCnt', 'objInfrstrPlaygrndCnt', 'objInfrstrSportGroundCnt', 'objInfrstrTrashAreaCnt',
                    'objInfrstrObjPrkngCnt' , 'objElevatorWheelchairCnt', 'objElevatorPassengerCnt', 'objElevatorCargoCnt',
                    'objElevatorCargoPassengerCnt','objInfrstrNotObjPrkngCnt', 'developer.problObjCnt', 'developer.buildObjCnt', 'developer.comissObjCnt']

    bool_type_columns = ['isAvailableWantLiveHere', 'objInfrstrRampFlg', 'objInfrstrCurbLoweringFlg',
                        'objInfrstrBicycleLaneFlg', 'objCompensFundFlg', 'objProblemFlg', 'objGuarantyEscrowFlg', 'objGreenHouseFlg',
                        'developer.fundGuarantyFlg', 'developer.govFundFlg']

    float_type_columns = ['objSquareLiving', self._price_column]

    date_type_columns = ['objReady100PercDt']

    df = self._as_type(df, int_type_columns, 'Int64')
    df = self._as_type(df, bool_type_columns, bool)
    df = self._as_type(df, float_type_columns, float)

    df['objReady100PercDt'] = pd.to_datetime(df['objReady100PercDt'], format='%Y-%m-%d')

    return df
```

```python
# класс миниисследования

class PlotVisualizer:

  def __init__(self):
    pass

  def show_sunburst(self, df, path, values):
    fig = px.sunburst(df, path=path, values=values, height=700, width=700)
    fig.show()

  def show_heatmap(self, values):
    plt.figure(figsize=(15, 15))

    ax = sns.heatmap(
      values,
      cmap='icefire',
      fmt='.0f',
      linewidths=.5
    )

  def show_hist(self, y, x, data):
    fig, ax = plt.subplots(figsize=(15, 15))

    ax = sns.barplot(
        y=y,
        x=x,
        data=data
        )

    ax.bar_label(ax.containers [ 0 ])
    plt.title(x)
    plt.show()

  def show_box(self, x, y, data):
     sns.catplot(
      x=x,
      y=y,
      data=data,
      kind='box',
      height=20
      );

     plt.title(x)
     plt.show()

  def show_corr_heatmap(self, matrix):
    plt.figure(figsize=(8, 6))
    sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('корреляция значений')
    plt.show()


class BuildingsAnalizer:

  def __init__(self, df, vizualizer):
    self._target_df = df
    self._vizualizer = vizualizer

    self._region = 'region'
    self._price = 'objPriceAvg'
    self._building_end_year = 'buildingYear'
    self._biuldings_end_date = 'objReady100PercDt'
    self._obj_square_living = 'objSquareLiving'
    self._flat_count = 'objFlatCnt'
    self._floor_count = 'objFloorCnt'
    self._parking_count = 'objElemParkingCnt'

    self._set_buildings_end_year()

    self._regions_group = self._target_df.groupby([self._region])
    self._regions_building_year_group = pd.DataFrame(self._target_df.groupby([self._region, self._building_end_year])[self._obj_square_living].sum())
    self._mean_by_regions = self._mean_by_regions()

  def get_price_mean(self):
    mean_price_regions = self._regions_group[self._price].mean().round(2)
    mean_price_regions.sort_values(inplace=True, ascending=False)

  def _set_buildings_end_year(self):
    self._target_df[self._building_end_year] =  self._target_df[self._biuldings_end_date].dt.year
    self._target_df[self._building_end_year].isna().sum()

  def vizualize_region_year_buildings_sunburst(self):
    self._vizualizer.show_sunburst(self._target_df, [self._building_end_year, self._region], self._obj_square_living)

  def vizualize_region_year_buildings_heatmap(self):
    buildings_by_years = self._regions_building_year_group.reset_index()
    buildings_by_years = buildings_by_years.pivot(self._region, self._building_end_year, self._obj_square_living)
    self._vizualizer.show_heatmap(buildings_by_years)

  def vizualize_stat_region_hist(self):
    plot_names = self._mean_by_regions.columns.to_list()
    feature_selector = self._get_feature_selector(plot_names)

    interact(
        self._plot_hist_graph,
        selected_feature=feature_selector
        )

  def vizualize_stat_region_box(self):
    plot_names = [self._price, self._flat_count, self._floor_count, self._obj_square_living, self._parking_count]
    feature_selector = self._get_feature_selector(plot_names)

    interact(
        self._plot_box_graph,
        selected_feature=feature_selector
        )

  def vizualize_corr_square_cnt(self):
    corr_df = self._target_df[[self._flat_count, self._floor_count, self._obj_square_living]]
    correlation_matrix = corr_df.corr()
    self._vizualizer.show_corr_heatmap(correlation_matrix)

  def _plot_hist_graph(self, selected_feature):
    self._mean_by_regions.sort_values(
        selected_feature,
        inplace=True,
        ascending=False
        )

    self._vizualizer.show_hist(
        self._mean_by_regions.index,
        selected_feature,
        self._mean_by_regions
        )

  def _plot_box_graph(self, selected_feature):
    self._vizualizer.show_box(
        selected_feature,
        self._region,
        self._target_df
        )

  def _mean_by_regions(self):
    func = 'mean'

    return (
    self._regions_group
    .agg(
          avg_price=(self._price, func),
          avg_flatts=(self._flat_count, func),
          avg_floors=(self._floor_count, func),
          avg_liv_area=(self._obj_square_living, func),
          avg_parking=(self._parking_count, func)
      )
    ).round(2)

  def _get_feature_selector(self, features):

    feature_selector = widgets.Dropdown(
        options=features,
        value=features[0],
        description='Выберите показатель'
    )

    return feature_selector
```
